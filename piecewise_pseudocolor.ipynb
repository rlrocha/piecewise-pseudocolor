# Transformação Piecewise e Pseudo Coloração

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Transformação piecewise
As variáveis e e s são a entrada e a saída da função piecewise, ou x e y como mostrado na aula de processamento pontual parte 2. A função alpha calcula o alfa de cada trecho, enquanto que a função beta calcula o beta, a partir de x, y e alfa de cada um dos trechos. O cálculo das funções segue a aula como base. As funções também serão utilizadas para pseudo coloração.

```python
def alpha(e1, e2, s1, s2):

    return (s2-s1)/(e2-e1)


def beta(e, s, a):
    
    return s-(a*e)
```

Definição e exibição das entradas e saídas da função piecewise. Os trechos são 0-100, 100-155, e 155-255 para entrada, para saída são 0-50, 50-205, 205-255 respectivamente. Os valores podem ser alterados sem nenhum problema para o melhor entendimento da transformação piecewise. Além disso, pode-se aumentar o número de trechos, porém precisam ser adicionados no loop da transformação piecewise.

```python
e = np.array([0, 100, 155, 255])
s = np.array([0, 50, 205, 255])

plt.plot(e, s)
plt.xlabel('Entrada')
plt.ylabel('Saída')
plt.title('Função Piecewise')
plt.grid()
plt.show()
```

Carregamento de imagem e loop para transformação da imagem usando piecewise.

```python
img_einstein = cv2.imread('images/einstein.tif', 0)

a1 = alpha(e[0], e[1], s[0], s[1])
a2 = alpha(e[2], e[1], s[2], s[1])
a3 = alpha(e[3], e[2], s[3], s[2])

b1 = beta(e[0], s[0], a1)
b2 = beta(e[1], s[1], a2)
b3 = beta(e[2], s[2], a3)

img_piecewise = np.zeros([img_einstein.shape[0], img_einstein.shape[1]])

for i in range(img_einstein.shape[0]):
  for j in range(img_einstein.shape[1]):
    # Primeiro trecho
    if img_einstein[i,j] <= e[1]:
      img_piecewise[i,j] = (img_einstein[i,j]*a1) + b1
    # Segundo trecho
    elif img_einstein[i,j] > e[1] or img_einstein[i,j] <= e[2]:
      img_piecewise[i,j] = (img_einstein[i,j]*a2) + b2
    # Terceiro trecho
    else:
      img_piecewise[i,j] = (img_einstein[i,j]*a3) + b3

img_piecewise = cv2.convertScaleAbs(img_piecewise)
```

A célula abaixo mostra os valores de alfa e beta para cada trecho. Os valores são iguais ou semelhantes aos apresentados na aula. A diferença se da devido ao arredondamento feito nos slides.

```python
print('Alfa 1: ', a1)
print('Beta 1: ', b1)
print('Alfa 2: ', a2)
print('Beta 2: ', b2)
print('Alfa 3: ', a3)
print('Beta 3: ', b3)
```

Exibição da imagem original e da imagem após a transformação piecewise. São utilizados vmin e vmax para mostrar a imagem antes e após a trasnformação, considerando que a imagem original possui baixo contraste, onde caso não fosse utilizado, o imshow iria ajustar a imagem entre 0 e 255 nos impedindo de visualizar as diferenças entre as imagens.

```python
fig_luz, axes = plt.subplots(1,2,figsize=(7,7), dpi = 100)
plt.tight_layout()

for ax in axes:
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

axes[0].imshow(img_einstein, vmin=0, vmax=255, cmap='gray')
axes[0].set_title('Imagem original')

axes[1].imshow(img_piecewise, vmin=0, vmax=255, cmap='gray')
axes[1].set_title('Transformação piecewise')

#plt.savefig('images/piecewise.png', bbox_inches = 'tight', dpi=300)
plt.show()
```

## Pseudo coloração

Para pseudo coloração utiliza-se o exemplo apresentado em aula, o qual é aplicada a pseudo coloração através de 3 cores com a transformação em trechos (piecewise). Entre o trecho 0-127 a cor será azul, em 127 a cor será branco e entre 128-255 a cor será vermelho. 

Desse modo, as entradas (x) serão 0, 127 e 255, porém agora terão 3 saídas, r, g e b, que representam os canais RGB. No ponto 0 (entrada ou x), o valor precisa ser azul segundo o exemplo do slide, desse modo, ele pode ser representado por r = 0, g = 0, e b = 255, que são representados pelas primeiras posições dos vetores ro, go, e bo, respectivamente. O mesmo acontece para o ponto 127, que precisa ser branco, ficando r = 255, g = 255, e b = 255. E por fim, o ponto 255, que deve ser representado por vermelhor, neste caso r = 255, g = 0, e b = 0.

Assim como na transformação piecewise, os valores podem ser alterados sem nenhum problema para melhor compreender o assunto. Além disso, pode-se adicinar outros trechos para representarem uma maior gama de cores, desde que sejam também adicionados no loop que realiza a pseudo coloração.

Os plots da célula abaixo representam os valores de entrada e suas respcetivas saídas para cada um dos canais RGB.

```python
e = np.array([0, 127, 255])

ro = np.array([0, 255, 255])
go = np.array([0, 255, 0])
bo = np.array([255, 255, 0])

fig_luz, axes = plt.subplots(1,3,figsize=(7,3), dpi = 100)
plt.tight_layout()

axes[0].plot(e, ro)
axes[0].set_title('Canal vermelho')
axes[0].set_xlabel('Entrada')
axes[0].set_ylabel('Saída')
axes[0].grid()
axes[1].plot(e, go)
axes[1].set_title('Canal verde')
axes[1].set_xlabel('Entrada')
axes[1].set_ylabel('Saída')
axes[1].grid()
axes[2].plot(e, bo)
axes[2].set_title('Canal azul')
axes[2].set_xlabel('Entrada')
axes[2].set_ylabel('Saída')
axes[2].grid()

plt.show()
```

A célula abaixo apresenta o loop que realiza a pseudo coloração. A e B são as matrizes que contém os valores de alfa e beta de cada trecho, e são utilizados no loop principal para realiazar a pseudo coloração.

```python
s = np.transpose(np.vstack((ro, go, bo)))

A = np.zeros([e.shape[0]-1, 3])
B = np.zeros([e.shape[0]-1, 3])

img_xray = cv2.imread('images/x_ray.jpg', 0)
img_pseu = np.zeros([img_xray.shape[0], img_xray.shape[1], 3])

for i in range(e.shape[0]-1):
    for j in range(3):
        A[i,j] = (s[i+1,j]-s[i,j])/(e[i+1]-e[i])
        B[i,j] = s[i+1,j]-(A[i,j]*e[i+1])

for i in range(img_xray.shape[0]):
    for j in range(img_xray.shape[1]):
      if img_xray[i,j] <= e[1]:
        # Primeiro trecho (0-127)
        img_pseu[i,j,0] = A[0,0]*img_xray[i,j] + B[0,0]
        img_pseu[i,j,1] = A[0,1]*img_xray[i,j] + B[0,1]
        img_pseu[i,j,2] = A[0,2]*img_xray[i,j] + B[0,2]
            
      elif img_xray[i,j] <= e[2] :
        # Segundo trecho (127-255)
        img_pseu[i,j,0] = A[1,0]*img_xray[i,j] + B[1,0]
        img_pseu[i,j,1] = A[1,1]*img_xray[i,j] + B[1,1]
        img_pseu[i,j,2] = A[1,2]*img_xray[i,j] + B[1,2]

img_pseu = cv2.convertScaleAbs(img_pseu)
```

A linha (posição 1) dos shapes abaixo representa a quantidade de trechos, 0-127 e 127-255. Já a coluna (posição 2) representa a quantidade de saídas (canais RGB).

```python
print(A.shape)
print(B.shape)
```

A célula abaixo apresenta um teste mais direto para verificar os valores de alfa e beta para cada um dos canais RGB. Quando a entrada (nível de cinza) é 0, a saída (canais RGB) é a cor azul (0, 0, 255). Se x é igual a 127, a cor é branco (255, 255, 255). E por fim, quando o valor do pixels é 255 a cor de saída da pseudo coloração é vermelho (255, 0, 0). 

```python
x = 127
y = np.zeros((3))

if x <= 127:
  y[0] = A[0,0]*x + B[0,0]
  y[1] = A[0,1]*x + B[0,1]
  y[2] = A[0,2]*x + B[0,2]
else:
  y[0] = A[1,0]*x + B[1,0]
  y[1] = A[1,1]*x + B[1,1]
  y[2] = A[1,2]*x + B[1,2]

print('Entrada: ', x)
print('Saída: ', y)
```

Exibição da pseudo coloração. A imagem original em escala de cinzas é transformada em imagem colorida através da pseudo coloração.

```python
fig_luz, axes = plt.subplots(1,2,figsize=(7,3), dpi = 100)
plt.tight_layout()

for ax in axes:
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

axes[0].imshow(img_xray, cmap='gray')
axes[0].set_title('Imagem original')

axes[1].imshow(img_pseu)
axes[1].set_title('Pseudo coloração')

#plt.savefig('images/pseudocolor.png', bbox_inches = 'tight', dpi=300)
plt.show()
```
